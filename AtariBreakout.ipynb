```python
import gymnasium as gym
import os
import gnwrapper

from stable_baselines3 import A2C
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_atari_env
```

## 2. Test Environment

```python
environment_name = "ALE/Breakout-v5"
#env = gym.make(environment_name, render_mode='rgb_array')
```

```python
env = gnwrapper.LoopAnimation(gym.make(environment_name, render_mode="rgb_array"))
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info, _ = env.step(action)
        score+=reward
    print(f'Episode:{episode} Score:{score}')
env.display()
env.close()
```

## 3. Vectorize Environment and Train Mode
- To vecrorize the environment and training with several different environments at once, to train the agents faster by training in parallel.

```python
# use helper function(make_atari_env) from stable baselines to create wrapped atari environments
env = make_atari_env("ALE/Breakout-v5", n_envs=4, seed=0)
```

```python
# stack the environments together
env = VecFrameStack(env, n_stack=4)
```

```python
env.reset()
```

```python
env.render()
```

```python
log_path = os.path.join('Training', 'ATARI', 'Logs')
model = A2C('CnnPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=500000)
```

## 4. Save and Reload Model

```python
a2c_path = os.path.join('Training', 'ATARI', 'Saved Models', 'A2C_500k_Breakout_Model')
model.save(a2c_path)
```

```python
del model
```

```python
model = A2C.load(a2c_path, env)
```

## 5. Evaluation and Test

```python
# the evaluation can not be done on multiple vectorize environments so need to vectorize it to a single env
env = make_atari_env("ALE/Breakout-v5", n_envs=1, seed=0)
env = VecFrameStack(env, n_stack=4)
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```

```python
env.close()
```

```python
obs = env.reset()
while True:
    action, _states = model.predict(obs)
    obs, rewards, done, info= env.step(action)
    env.render()
env.close()
```
